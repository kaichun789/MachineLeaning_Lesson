# 非線形回帰問題 y=sin(x)の学習

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

#サンプル数
num_of_sam = 200
max_of_sam = 2*np.pi
x = np.random.uniform(0,max_of_sam,num_of_sam)
noise = np.random.normal(0,0.1,num_of_sam)
#正解データの生成
t = np.sin(x)+noise

#グラフ描画
plt.scatter(x,t,s=5)
plt.title("Correct data",fontsize=25)
plt.xlabel("Input(x)",fontsize=18)
plt.ylabel("Output(y)",fontsize=18)
plt.show()
```

```python
#シグモイド関数を使えるようにする
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

#隠れ層のニューロン数
dimention = 20

#重みの生成
W1 = np.random.randn(dimention,1)
W2 = np.random.randn(1,dimention)

#バイアスの生成
B1 = np.random.randn(dimention,1)
B2 = np.random.randn(1,1)

learning_rate = 0.0002
E_save = np.array([])
num_of_itr = 10000

#学習
for i in range(num_of_itr):
    #順伝番 forward propagation
    H = sigmoid(x*W1+B1)
    y = np.dot(W2,H)+B2
    E = np.sum((t-y)**2)
    E_save = np.append(E_save,E)

    #differential
    dW2 = 2*np.sum(H*(y-t),axis=1)
    dB2 = 2*np.sum(y-t)
    dW1 = 2*W2*np.sum(x*H*(1-H)*(y-t),axis=1)
    dB1 = 2*W2*np.sum(H*(1-H)*(y-t))

    #逆伝番 back propagation
    W1 = W1-learning_rate*dW1.T
    W2 = W2-learning_rate*dW2
    B1 = B1-learning_rate*dB1.T
    B2 = B2-learning_rate*dB2
    
X_line = np.linspace(0, max_of_sam, 200)
H_line = sigmoid(X_line*W1+B1)
Y_line = np.ravel(np.dot(W2,H_line)+B2)
```

```python
plt.scatter(x, t,s=5,label="Correct")
plt.plot(X_line, Y_line,"red",label="Learning Result")
plt.title("Learning Result",fontsize=25)
plt.xlabel("Input(x)",fontsize=18)
plt.ylabel("Output(y)",fontsize=18)
plt.legend()
plt.show()
```
